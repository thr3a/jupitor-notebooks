```python
dummies = pd.get_dummies(pokemons[pokemons.columns])
dummies.head()
```

```python
from sklearn.model_selection import train_test_split
x = dummies.drop(['Legendary'], axis=1).loc[:, 'Total':'Type 2_Water']
y = dummies['Legendary'].values
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=151)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
estimate_num = 100
forest = RandomForestClassifier(random_state=22, n_estimators=estimate_num, n_jobs=-1)
```

```python
forest.fit(X_train, y_train)
```

```python
forest.score(X_train, y_train)
```

```python
forest.score(X_test, y_test)
```
